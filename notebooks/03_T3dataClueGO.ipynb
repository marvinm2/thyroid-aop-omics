```python
%pip install py4cytoscape
import py4cytoscape as p4c
import os
import pandas as pd


# Initialize Cytoscape connection
p4c.cytoscape_ping()


```

```python
# Load hypothyroid dataset
hypothyroid_df = pd.read_csv('../data/processed/hypothyroid_condition.csv', sep='\t')
```

```python
# Filter for significant genes only, then create nodes
sig_genes = hypothyroid_df[hypothyroid_df['Significance'] == 1].copy()
print(f"Significant genes: {len(sig_genes)}")

nodes = sig_genes[['GeneSymbol', 'log2FC', 'Significance']].copy()
nodes['id'] = nodes['GeneSymbol']  # Assigning ID as GeneSymbol
```

```python
# Convert to list of dictionaries for Cytoscape import
nodes_list = nodes.to_dict('records')
```

```python
# Create a network in Cytoscape (without applying layout - will use ClueGO later)
network_suid = p4c.create_network_from_data_frames(nodes=nodes, title='Hypothyroid Network', collection='RNA-Seq Networks')

# Skip layout since ClueGO will be used
print(f"Network created with SUID: {network_suid}")
```
